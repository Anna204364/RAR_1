Транспортная задача

```python
#Введем значения объёма продукции (в кг.) на трех складах
a1 = float(input("Объёма продукции склада 1: "))
a2 = float(input("Объёма продукции склада 2: "))
a3 = float(input("Объёма продукции склада 3: "))

#Введем значения потребности трех магазинов в продукции (в кг.)
b1 = float(input("Потребность магазина 1: "))
b2 = float(input("Потребность магазина 2: "))
b3 = float(input("Потребность магазина 3: "))

from cvxopt.modeling import variable, op
import time
start = time.time() 

#Variable - векторная переменная 
#Первый аргумент — размерность вектора. Второй аргумент — это строка с именем переменной
x = variable(9, 'x')

#Матрица значений стоимости
c = list(map(float,input('Введите значения цен через пробел ').split()))

#Целевая функция - общая стоимость перевозки
z = (c[0]*x[0] + c[1]*x[1] + c[2]*x[2] + c[3]*x[3] + c[4]*x[4] + c[5]*x[5] + c[6]*x[6] + c[7]*x[7] + c[8]*x[8]) 

#Ограничения
#Мы не можем перевезти товара больше, чем имеется на складе
mass1 = (x[0] + x[1] + x[2] <= a1)
mass2 = (x[3] + x[4] + x[5] <= a2)
mass3 = (x[6] + x[7] + x[8] <= a3)

#Перевозить будем ровно столько, сколько необходимо потребителям
mass4 = (x[0] + x[3] + x[6] == b1)
mass5 = (x[1] + x[4] + x[7] == b2)
mass6 = (x[2] + x[5] + x[8] == b3)

x_non_negative = (x >= 0)   

#По переданым в качестве параметров целевой функции переменным и набору ограничений создает объект модели
problem = op(z, [mass1, mass2, mass3, mass4, mass5, mass6, x_non_negative])
problem.solve(solver = 'glpk')  
print("Матрица:")
for i in x.value:
         print(i)
print("Стоимость доставки:")
print(problem.objective.value()[0])
stop = time.time()
print ("Время :")
print(stop - start)
```
